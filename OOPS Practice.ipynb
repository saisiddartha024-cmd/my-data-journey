```python
class Circle():

    #class object atttibute
    pi = 3.14

    def __init__(self,radius=1):

        self.radius = radius

    #method
    def get_circumfrence(self):
        return self.radius * self.pi *2
```

```python
my_circle = Circle(30)
```

```python
my_circle.pi
```

```python
class Animal():

    def __init__(self):
        print("Animal Created")

    def who_am_i(self):
        print("I am an animal")

    def eat(self):
        print("am eating")
```

```python
class Dog(Animal):

    def __init__(self):
        Animal.__init__(self)
        print("dog created")
    def who_am_i(self):
        print("iam a dog")
    def eat(self):
        print("ama dog  eating")

    def bark(self):
        print("woof")
```

```python
mydog= Dog()
```

```python
mydog.who_am_i()
```

```python
mydog.bark()
```

```python
class Animal():
    def __init__(self,name):
        self.name = name

    def speak(self):
        raise NotImplementedError("sub class must implement tgis abstract method")
```

```python
class Dog(Animal):

    def speak(self):
        return self.name+ "says woof"
    
```

```python
class Cat(Animal):

    def speak(self):
        return self.name+ "says meow"
```

```python
fido = Dog("fido")
```

```python
isis = Cat("isis")
```

```python
print(fido.speak())
```

```python
print(isis.speak())
```

```python
#magic methods
class Book():

    def __init__(self,title,author,pages):

        self.title =title
        self.author =author
        self.pages =pages

    def __str__(self):
        return f'{self.title} by {self.author}'

    def __len__(self):
        return self.pages

    def __del__(self):
        print("A book object deleted")
```

```python
b = Book('python rocks','siddartha',200)
```

```python
print(b)
```

```python
str(b)
```

```python
len(b)
```

```python
del b
```
