```python
# Day 2 – Lists Mastery – 9 Dec 2025
# Sai Siddartha

fruits = ["apple", "banana", "orange", "mango"]
print("Original:", fruits)

print("Third fruit:", fruits[2])
print("Last fruit:", fruits[-1])
print("Middle two:", fruits[1:3])

fruits.append("grape")
print("After append:", fruits)

fruits[0] = "WATERMELON"
print("After changing first:", fruits)

numbers = [10, 20, 30, 40, 50]
numbers.pop()
print("After pop:", numbers)

mixed = [1, "Sai", True, 3.14, ["nested", "list"]]
print("Mixed list:", mixed)
print("Nested item:", mixed[4][1])
```
