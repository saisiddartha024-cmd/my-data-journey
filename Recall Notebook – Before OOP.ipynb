```python
# Recall Notebook – Before OOP
# Sai Siddartha Thota | 17 Dec 2025
# Quick refresh of everything from Sections 1–7

print("=== VARIABLES & TYPES ===")
name = "Sai"
age = 23
hours_today = 5.5
is_learning = True

print(f"Name: {name}, Age: {age}, Hours: {hours_today}, Learning: {is_learning}")

print("\n=== LISTS ===")
skills = ["Python", "GitHub", "Loops", "Functions"]
print("Skills list:", skills)
print("First skill:", skills[0])
skills.append("OOP soon")
print("After append:", skills)

print("\n=== DICTIONARIES ===")
progress = {"days_coding": 9, "sections_done": 7, "goal": 600}
print("Progress dict:", progress)
print("Sections done:", progress["sections_done"])
progress["hours"] = 150
print("Updated:", progress)

print("\n=== TUPLES (immutable) ===")
coordinates = (51.5074, -0.1278)  # London
print("London coords:", coordinates)
# coordinates[0] = 40.7128  # would error — immutable

print("\n=== SETS (unique items) ===")
unique_skills = {"Python", "SQL", "Python"}  # duplicate removed
print("Unique skills:", unique_skills)

print("\n=== FOR LOOP ===")
for skill in skills:
    print(f"Learning: {skill}")

print("\n=== WHILE LOOP ===")
count = 1
while count <= 3:
    print(f"Day {count} complete")
    count += 1

print("\n=== IF / ELIF / ELSE ===")
hours = 6
if hours >= 7:
    print("Legendary day!")
elif hours >= 5:
    print("Strong day — keep it up")
else:
    print("Good start")

print("\n=== FUNCTIONS ===")
def motivate(hours):
    return f"You studied {hours} hours — proud of you!"

print(motivate(6))

print("\n=== LIST COMPREHENSION ===")
squares = [x**2 for x in range(1, 6)]
print("Squares 1-5:", squares)

even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]
print("Even squares:", even_squares)

print("\nAll good — ready for classes and objects!")
```
