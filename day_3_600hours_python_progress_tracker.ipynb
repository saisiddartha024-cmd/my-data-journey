```python
# Day 3 – Python Progress Tracker
# Sai Siddartha Thota | 10 Dec 2025
# A simple tool to track my daily study hours and predict job readiness

# My study log (hours per day)
study_hours = [5, 6, 4.5, 7, 6, 5.5, 8]   # last 7 days

# Target for junior data analyst job
target_hours = 600   # total hours needed (estimate)

# Calculate progress
total_hours_so_far = sum(study_hours)
days_studied = len(study_hours)
average_hours_per_day = total_hours_so_far / days_studied
hours_remaining = target_hours - total_hours_so_far
estimated_days_left = round(hours_remaining / 6)  # assuming 6 hrs/day from now

# Progress percentage
progress_percent = (total_hours_so_far / target_hours) * 100

# Decision logic
print("=== DAILY PYTHON STUDY TRACKER ===")
print(f"Total hours studied     : {total_hours_so_far}")
print(f"Average per day          : {average_hours_per_day:.1f} hours")
print(f"Progress to 600h goal   : {progress_percent:.1f}%")
print()

if progress_percent >= 100:
    print("GOAL ACHIEVED — Ready for interviews!")
elif progress_percent >= 70:
    print("Strong progress — keep pushing!")
elif progress_percent >= 50:
    print("Halfway there — momentum building!")
else:
    print("Early days — but started strong!")

print(f"Estimated days left     : ~{estimated_days_left} days")
print("Keep going. No excuses.")
```
