```python
# Day 5 â€“ For Loops & List Comprehensions
# Sai Siddartha Thota | 12 Dec 2025

# Basic for loop through a list
cities = ["London", "Manchester", "Birmingham", "Leeds", "Glasgow"]

for city in cities:
    print(f"Processing {city} data...")

print()  # just a blank line

# Loop with range
for day in range(1, 6):
    print(f"Day {day} of consistent coding")

print()

# Adding up numbers in a list
hours = [5, 6, 7, 4, 8]
total = 0
for h in hours:
    total += h
print(f"Total study hours last 5 days: {total}")

print()

# Loop with enumerate (index + value)
skills = ["Python", "SQL", "GitHub", "Pandas", "Power BI"]
for index, skill in enumerate(skills, start=1):
    print(f"{index}. Learning {skill}")

print()

# Nested loops example
print("--- Mini daily plan ---")
for day in range(1, 4):
    print(f"Day {day}:")
    for task in ["Wake up", "Code", "Work", "Code again"]:
        print(f"  - {task}")

print()

# List comprehensions (the cool one-liner way)
# Normal way
squares = []
for x in range(10):
    squares.append(x**2)
print("Normal squares:", squares)

# Pro way
squares_pro = [x**2 for x in range(10)]
print("One-liner squares:", squares_pro)

# With condition
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print("Even squares only:", even_squares)

# Quick test I tried
names = ["sai", "rahul", "aisha"]
upper_names = [name.title() for name in names]
print("Capitalised names:", upper_names)

print("\nFeeling good with loops now!")
```
