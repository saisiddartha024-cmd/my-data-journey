```python
# Day 6 – While Loops + break, continue, pass
# Sai Siddartha Thota | 13 Dec 2025 (late night grind)

print("=== BASIC WHILE LOOP ===")
count = 1
while count <= 5:
    print(f"Day {count} — still coding")
    count += 1

print("\n=== WHILE WITH BREAK ===")
target = 100
hours = 0

while hours < target:
    hours += 20
    print(f"Hours so far: {hours}")
    if hours >= 80:
        print("Good progress — stopping early with break")
        break

print("\n=== WHILE WITH CONTINUE ===")
day = 0
while day < 7:
    day += 1
    if day == 4:          # skip rest day
        print("Day 4: Rest day — continue to next")
        continue
    print(f"Day {day}: Coding session")

print("\n=== WHILE WITH PASS ===")
attempt = 0
while attempt < 3:
    attempt += 1
    if attempt == 2:
        pass              # placeholder — do nothing
        print("Attempt 2: pass used — just testing")
    print(f"Attempt {attempt} complete")

print("\nWhile loops + control statements understood.")
```
